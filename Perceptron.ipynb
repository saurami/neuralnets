The input data is parsed into a matrix in which each occurrence of `#` is changes to 1 and that of `.` is changed to 0.
In this case, each letter of the dataset is represented by its respective list of ones and zeroes. There is one list which consists of 26 sublists.

```python
from dataset import *

training_data=[]
for row in TRAINING_DATA:
    training_row=[]
    for column in row:
        for element in column:
            val=int(element.replace('.','0').replace('#','1'))
            training_row.append(val)
    training_data.append(training_row)
```

The `fit` method is used to train the perceptron and adjust the values of weights and biases for the given training set.

The `predict` method is used to estimate the output based on the given input features having weights and bias. If there is a match, the estimate output is 1 and if not the out is 0.

```python
import numpy as np

class Perceptron(object):

    def __init__(self, learning_rate=0.01, epochs=15):
        self.learning_rate = learning_rate
        self.epochs = epochs

    def train(self, x, y):
        self.weights = np.random.rand(1+ x.shape[1])
        self.errors = []

        for _ in range(self.epochs):
            errors_per_epoch = 0
            for x_input, target in zip(x, y):
                change = self.learning_rate * (target - self.predict(x_input))
                self.weights[1:] += change * x_input
                self.weights[0] += change
                errors_per_epoch += int(change != 0.0)
            self.errors.append(errors_per_epoch)
        return self

    def predict(self, x):
        return np.where((np.dot(x, self.weights[1:]) + self.weights[0]) >= 0.0, 1, 0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

X_matrix = np.array(training_data)
Y_matrix = np.eye(26)
```
